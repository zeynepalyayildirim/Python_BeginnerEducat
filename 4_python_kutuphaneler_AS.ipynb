#Kütüphaneler

---

Kütüphaneler, daha önceden yapılmış işlemleri saklayan ve bunları tekrar kullanabilmemizi sağlayan bir yapıdır. Kütüphaneler ücretsiz bir şekilde kullanılabilir olabileceği gibi paralı da olabilir. Bunları yazan kişiler, tekrar tekrar çağırabilmemiz için erişime açık isimler kullanarak oluştur muşlardır.Milyarlarca kütüphane vardır. Örneğin: <br>

* Rastgele sayı veya içerik üretmemizi sağlayan **random** kütüphanesi
* İşletim sisteminde yer alan klasörlerle işlem yapmamızı sağlayan **os** kütüphanesi
* Matematiksel üst düzey işlemler yapmamızı sağlayan **math** kütüphanesi
* Oyun geliştirmek için **turtle** ve **pygame** kütüphaneleri
* Arayüzlü program tasarlamamızı sağlayan **tkinter** kütüphanesi
* Veritabanı araçlarını kullanabilmeyi sağlayan **sqlite** kütüphanesi

##Peki Kütüphaneler Nasıl Kullanılır?
Kütüphaneleri kullanmak için öncelikle kod metnimizin en tepesinde onları çağırmamız gerekir. Eğer bir kütüphane bilgisayarımızda yüklü ise kütüphane hatasız çalışır. Kütüphaneyi kodumuzu eklemek için **import** deyimini kullanıyoruz. Örneğin: <br>

```
import random
import os
import math
```








````python
#Kütüphaneleri programın başında çağırıyoruz.
import random #Rastgele içerik üretmeyi sağlar.

rastgele_sayi = 0

#Rastgele sayıyı random kütüphanesinden çağırmak istiyorsam, kütüphanenin adını yazarım.
#Ardından random nokta işareti ile istediğim işlemi çağırırım.
rastgele_sayi = random.randint(1,5)
#randint-> girilen iki değer arasında rastgele sayı üretmemizi sağlar.
print(rastgele_sayi)
if rastgele_sayi == 1:
  print("Uğur")
elif rastgele_sayi == 2:
  print("Zeynep")
elif rastgele_sayi == 3:
   print("Müfit")
elif rastgele_sayi == 4:
  print("Ataberk")
if rastgele_sayi == 5:
  print("Yiğit")
````

**NOT:** Kütüphaneleri çağırırken kısaltma isimlerel de çağırabiliriz.
Bunun için **as** takısı kullanılır.
**Örneğin;** import randon as r

````python
import random as rd #Kısaltma yapılabiliyor.
import math as m

sayi1 = rd.randint(1,10) #1-10 arasında rastgele sayıyı değişkene atabiliyoruz.
print(sayi1)

#MATEMATİKSEL FONKSİYONLAR
print(m.factorial(sayi1)) #Faktöriyel
print(m.comb(sayi1, 2)) #Kombinasyon (İki değer alır) (sayi1,sayi2)
print(m.perm(sayi1, 2)) #Permutasyon
print(m.sqrt(sayi1)) #Karekök alma
print(m.floor(sayi1)) #Yuvarlama
print(m.pow(sayi1, 4)) #İstediğin sayının kökkuvvetini alma
print(m.log(sayi1, 3)) #İstediğin tabandan logaritma alma
print(m.pi) #Pi sayısı
print(m.e) # e sayısı
````
