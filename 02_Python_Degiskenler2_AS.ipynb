#Değişkenlerle İşlemler
String yani metin değişkenlerinin kendilerine özgü işlemleri, sayıların kendine özgü ayrı işlemleri vardır.İlk olarak sayı değişkenleri ile işlemlere başlayalım.

**NOT:** "print"'te + işareti aynı türden değerleri yan yana yazar veya toplar.
,(virgül) işareti farklı türden nesneleri ayrı olarak algılar ve yan yana yazabilmemizi sağlar.

```python
#Sayı değişkenini işleme sokma;
birinci_sayi = int(input("Birinci Sayıyı Giriniz: "))
ikinci_sayi = int(input("İkinci Sayıyı Giriniz: "))
#NOT: Genelde bir işlem yapılıyorsa o işlemi ayrı bir değişkene yazarız.
#NOT-2: İnput'u int() parantezine alarak yazabiliriz. int(input(vs))
toplam = birinci_sayi + ikinci_sayi
cikarma = birinci_sayi - ikinci_sayi
carpma = birinci_sayi * ikinci_sayi #Çarpma işareti yıldızdır.
bolme = birinci_sayi / ikinci_sayi
print(int(birinci_sayi),"ile",int(ikinci_sayi),"arasında 4 işlem şu şekildedir:")
print("Toplam:",int(toplam))
print("Çıkarma:",int(cikarma))
print("Çarpma:",int(carpma))
print("Bölme:",int(bolme))
```
