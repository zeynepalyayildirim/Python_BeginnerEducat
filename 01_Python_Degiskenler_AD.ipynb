#DEĞİŞKENLER
Yazılımda içinde veri sakladığımız yapılara değişken diyoruz.Bu veri yazılımcının veya kullanıcının saklayabileceği önemli bilgilerdir.

```python
#Bir değişken oluşturalım.
#Değişkenim benim adımı soyadımı saklasın.
#Bir değişken şöyle oluşturulur;
#DeğişkeninAdı = Değişkenin değeri(içinde  sakladığı bilgi)
#1.YOL
AdSoyad = "Zeynep Yıldırım"
#2.YOL
Ad = "Zeynep"
Soyad = "Yıldırım"
#Sizce hangi yöntem daha doğrudur?
#Her ikisi de doğru, daha detaylı bilgi almak istersem ikinci yöntem ama daha basit olsun dersem birinci yöntemi seçerim.
print(AdSoyad)
print(Ad+" "+Soyad)
```

#Değişken oluştururken Dikkat!

1.  Değişken ismini değişkenin amacına göre içinde barındırdığı bilgiye göre vermemiz gerekir.
2.  Değişken isimlerini oluştururken Türkçe Karakter kullanmıyorum.(**Türkçe Karakterler: ı,ç,ö,ü,ş,ğ,İ**)
3.  Değişken oluşturuken isimler arasına boşluk bırakmadan  yazarız.
Ad Soyad (**Yanlış**)
Ad_Soyad (**Doğru**)

```python
#Değişken Türleri
#Bir Arabanın Özellikleri
araba_marka = "Tesla" #Metin -> String (str) demektir.
araba_hiz = 4000 #Sayi -> Integer (int) demektir.
arabanın_yüksekligi = 139.9 #Yukseklik -> Float (f) demektir.
arabanin_genisligi = 250,3 #Genişliği -> Double (d) demektir.
arabanin_uzunlugu = 390.2 #Float kullanilir.
araba_calisiyor_mu = True #Doğru/Yanlış -> Boolean (bool) demektir.
araba_pahali_mi = False #Doğru = True, Yanlış = False,
#Tek print'te ama alt alta bütün değişkenlerin değelerini nasıl yazarım?
print(araba_marka+"\n"+str(araba_hiz)+"\n"+str(arabanın_yüksekligi)+"\n"+str(arabanin_genisligi)+"\n"+str(araba_calisiyor_mu)+"\n"+str(araba_pahali_mi))
```

#Değişkenlerle İşlemler ❤
Değişkenleri içindeki değerden bağımsız olarak işleme sokabiliriz.Sayıları toplayabilir, metinleri yan yana veya alt alta yazdırabiliriz.İçindeki değer amacı doğrultusunda kullanılmalıdır! Değişkenleri işleme sokmak için yeni bir değişken de oluşturulabilir. 🤔

```python
sayi1 =23
sayi2 =38
toplam = sayi1+sayi2
print(toplam)
print(str(sayi1)+"+"+str(sayi2)+"="+str(toplam))
```
