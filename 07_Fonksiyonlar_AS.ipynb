Farklı proglamlama dillerinde metotlar olarak biline fonksiyonlar, nesne tabanlı proglamlamanın temelidir."Nesne Tabanlı Proglamlama(NTP)" İngilizce "Object Oriented Programming (OOP)" olarak bilinir. Bu proglamlama (NTP) yazılımda tekrar eden veya sonradan kullanılabilen işlemleri ve değişkenleri saklamayı amaçlar. <br>

**Örneğin;** <br>
1. Bir listeyi birden fazla kez kullanmak istiyorsunuz diyelim. NTP bunu bir kez yazıp, aşağıdaki kodlarda tekrar cağırabilmemizi sağlar.
2. Bir matematik işlemi yapıyoruz diyelim. Bu işlemi bir fonksiyonun içine saklayıp, o fonksiyonu çağırarak satırlarca kod yazmamıza engel olur.
3. Bir programda yazdığımız işlemleri başka bir programa yükleyerek veya dosyalarına ekleyerek onu kullanabiliriz. Bu işlemi import ile yaparız. Tıpkı kütüphane ekler gibi kendi kütüphanelerimizi oluşturabiliriz.

````python
#Parametre parantez içine yazılır.
def kullanici_paneli(kullanici_adi,parola):
  if kullanici_adi == "Gary" and parola == "12A34B":
    print("Hoşgeldiniz")
  else:
    print("Hatalı Giriş")
````

#Değer Döndüren ve Değer Döndürmeyen Fonksiyonlar

Bizim oluşturduğumuz daha önceki fonksiyonlar değer döndürmeyen fonksiyonlardı.Yani işlemi yaptırır ve sadece bir kod satırında çağırarak çalışırlardı. Ancak **değer döndüren** fonksiyonlar ise belirli bir tür değişkenin içerisine veri aktarılarak çalışan fonksiyonlardır.

##Peki, Değer Döndüren Fonksiyon Ne Demek Oluyor?
Değer döndürmeyen fonksiyon bir değişkenin içerisine değer atılarak yazılır. Bu değişken fonksiyonda belirlenen aynı türden değişken olmalıdır. Yani bir fonksiyonun sonucunda değişkene sadece bir değer atılır. Bu değer string ise string türünde, integer ise integer türünde bir değişkenin içine yazılır.

````python
#Değer Döndürmeyen Fonksiyon
def toplama(sayi1,sayi2):
  return sayi1+sayi2
  print("Toplam:",toplama)
````

````python
#Değer Döndüren Fonksiyon
def toplam(sayi1,sayi2):
  return sayi1+sayi2 #Return komutunu kullanıyorum.

#ANA PROGRAM
a = int(input("Sayı Giriniz:"))
b = int(input("Sayı Giriniz:"))
sonuc = toplam(a,b)) #Bir değişken içerisine yazıyorum
print(sonuc)
````

#OS KÜTÜPHANESİYLE DOSYA İŞLEMLERİ YAPMA

OS kütüphanesi "Operating System" yani "İşletim Sistemi" anlamına gelmektedir. Dosya açma, oluşturma, klasör yollarını öğrenme, klasör oluşturma gibi işletim sisteminizde gezinmenizi sağlayan bir kütüphanedir.OS kütüphanesinde yaygın kullanılan open fonksiyonu vardır.

**NOT:** Oluşturacağınız dosyanın uzantısını bilmeniz gerekir.
Word dosyasının uzantısı .docx
Not defterinin uzantısı .txt
Örnek bir resim dosyası .png olabilir.

os.open -> Bir dosya aç demek. Bu dosya klasör veya herhangi bir dosya olabilir. <br>
w -> write yani dosyaya yaz demek.<br>
r -> read sadece oku demek. <br>
a -> all yani hem yaz hem oku.
